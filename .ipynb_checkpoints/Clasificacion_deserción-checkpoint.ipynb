# 5. Limpieza de datos y division del set de datos para el entrenamiento y prueba

```python
# Separacin de las variables numericas y no numericas
df_numericas = df_raw.select_dtypes(include=np.number)
df_numericas.columns
```

```python
df_catagorical  =df_raw.select_dtypes(exclude=np.number)
df_catagorical.columns
```

```python
print(" Total columnas antes de la divisin: ", df_raw.shape[1],
     "\n Total columnas numericas:            ", df_numericas.shape[1],
     "\n Total columnas categoricas:          ", df_catagorical.shape[1],
     "\n Suma de columnas:                    ", df_numericas.shape[1] + df_catagorical.shape[1])
```

```python
# Transformacon de la data de tipo categrica  a numrica 
from sklearn.preprocessing import OneHotEncoder
encoderOneHot = OneHotEncoder()
df_catagorical = encoderOneHot.fit_transform(df_catagorical).toarray()
df_catagorical
```

```python
df_catagorical.shape
```

```python
df_catagorical = pd.DataFrame(df_catagorical)
df_catagorical
```

```python
df_numericas
```

```python
# Union de los sets preparados
df_concatenado = pd.concat([df_catagorical, df_numericas], axis= 1)
# Se almacena la variable objetivo en y
y = df_concatenado['Attrition']
# Se elimina del set la variable objetivo
df_final = df_concatenado.drop('Attrition', axis= 1)
df_final.head()
```

```python
# Escalacion de valores de los datos (mnimo 0 y mximo 1)
from sklearn.preprocessing import MinMaxScaler
minMaxscalare = MinMaxScaler()
```

```python
df_scaled = minMaxscalare.fit_transform(df_final)
df_scaled
```

## 6. Aplicacin de los modelos

```python
# Preparacin y division  de la data entre el set de entrenamiento y prueba
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(df_scaled, y,
                                                    test_size=0.25 )
```

```python
print("X_train.shape: ", X_train.shape, "\nX_test.shape", X_test.shape)
```

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(f"Accuracy: {100* accuracy_score(y_test, y_pred) }%")
```

```python
# Matriz de confusion 
# Resultados en el Conjunto de Testing para el modelo de Logistic Regression
cm_lr = confusion_matrix(y_test, y_pred)
sns.heatmap(cm_lr, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)
cm_rfc = confusion_matrix(y_test,y_pred_rfc )
```

```python
sns.heatmap(cm_rfc, annot=True)
```

```python
print(classification_report(y_test,y_pred_rfc ))
```

```python
X_train.shape
```

```python
# Creacin de una red neuronal profunda
rneuronal = tf.keras.Sequential()
rneuronal.add(tf.keras.layers.Dense(49, activation= 'relu', input_shape=(49,) ))

rneuronal.add(tf.keras.layers.Dense(500, activation= 'relu' ))
rneuronal.add(tf.keras.layers.Dense(256, activation= 'relu' ))

rneuronal.add(tf.keras.layers.Dense(1, activation= 'sigmoid' ))
```

```python
rneuronal.summary()
```

```python
rneuronal.compile(optimizer='Nadam',
                    loss= "binary_crossentropy",
                    metrics=['accuracy'],)
```

```python
epoch_hist = rneuronal.fit(X_train, 
                           y_train, 
                           epochs=17)
```

```python
y_pred_rneuro = rneuronal.predict(X_test)
y_pred_rneuro
```

```python
y_pred_rneuro_umbral = y_pred_rneuro > 0.6
y_pred_rneuro_umbral
```

```python
epoch_hist.history.keys()
```

```python
plt.plot(epoch_hist.history['loss'])
plt.plot(epoch_hist.history['accuracy'])
plt.title("Funcion perdida del modelo durante el Entrenamiento")
plt.xlabel("Epochs")
plt.ylabel("Error del entrenamiento")
plt.legend(['loss','accuracy'])
```

```python
cm_neural = confusion_matrix(y_test, y_pred_rneuro_umbral)
sns.heatmap(cm_neural, annot=True)
```
